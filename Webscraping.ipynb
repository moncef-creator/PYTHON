# **Webscraping DATA From Table**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')

findsup=soup.find_all('table')[0]
world_title=findsup.find_all('th')
world_title_table= [title.text.strip() for title in world_title]

df= pd.DataFrame(columns = world_title_table)
df
```

```python
rows = findsup.find_all('tr')
```

```python
for Cellules in rows [1:] :
    row_data = Cellules.find_all('td')
    DATAS = [data.text.strip() for data in row_data]
    # print(DATAS)
    lenght = len(df)
    df.loc[lenght]= DATAS
```
